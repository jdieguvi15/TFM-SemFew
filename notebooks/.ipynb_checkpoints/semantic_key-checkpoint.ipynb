```python
labels = ['baby', 'bed', 'bicycle', 'chimpanzee', 'fox', 'leopard', 'man',
       'pickup_truck', 'plain', 'poppy', 'rocket', 'rose', 'snail',
       'sweet_pepper', 'table', 'telephone', 'wardrobe', 'whale', 'woman',
       'worm']
```

```python
# Use wordnet for small definition
import nltk
from nltk.corpus import wordnet as wn
from transformers import pipeline

def get_wordnet_definition(class_name):
    synsets = wn.synsets(class_name)
    if synsets:
        return synsets[0].definition()  # TODO: Take the first definition?
    return None
```

Original Prompt:

f'"{definition}" is the definition of the {class_name}. Please rewrite and expand this definition to make it more detailed and consistent with scientific fact. Briefness is required, using only one paragraph.'

Proposed Prompt:

f'"{definition}" is the definition of the {class_name}. Please rewrite and expand this definition to make it more visually detailed and consistent with scientific fact for a IA model to differentiate it from this class to other classes like {set(labels) - set([class_name])}. Briefness is required, using only one paragraph.'

```python
from google import genai
from openai import OpenAI
from groq import Groq

def improve_definition(class_name, definition, prompting='Original', use="Gemini"):
    if prompting == 'Original':
        prompt = f'"{definition}" is the definition of the {class_name}. Please rewrite and expand this definition to make it more detailed and consistent with scientific fact. Briefness is required, using only one paragraph.'
    else:
        prompt = f'"{definition}" is the definition of the {class_name}. Please rewrite and expand this definition to make it more visually detailed and consistent with scientific fact for a IA model to differentiate it from this class to other classes like {set(labels) - set([class_name])}. Briefness is required, using only one paragraph.'

    if use == "Gemini":
        client = genai.Client(api_key=GENAI_KEY)
    
        response = client.models.generate_content(
            model="gemini-2.0-flash", contents=prompt
        )
        answer = response.text
        
    elif use == "DeepSeek":
        client = OpenAI(
          base_url="https://openrouter.ai/api/v1",
          api_key=OPENAI_KEY,
        )
        
        completion = client.chat.completions.create(
          extra_body={},
          model="deepseek/deepseek-r1:free",
          messages=[
            {
              "role": "user",
              "content": prompt
            }])
        answer = completion.choices[0].message.content

    elif use == "Groq":
        client = Groq(api_key=GROQ_KEY)
        chat_completion = client.chat.completions.create(
            messages=[
                {
                    "role": "user",
                    "content": prompt,
                }
            ],
            model="llama3-8b-8192",
        )
        answer = chat_completion.choices[0].message.content
        
    return answer
```

```python
for class_name in labels:
    print(class_name)
    definition = get_wordnet_definition(class_name)
    print("WordNet defintion: \n", definition, "\n")
    semantic = improve_definition(class_name, definition)
    print("Gemini Semantics: \n", semantic)
    semantic2 = improve_definition(class_name, definition, use="DeepSeek")
    print("DeepSeek Semantics: \n", semantic2)
    semantic3 = improve_definition(class_name, definition, use="Groq")
    print("Groq Semantics: \n", semantic3)
    print("\n")
    break
```

```python
for class_name in labels:
    print(class_name)
    definition = get_wordnet_definition(class_name)
    print("WordNet defintion: \n", definition, "\n")
    semantic = improve_definition(class_name, definition, prompting='')
    print("Gemini Semantics: \n", semantic)
    semantic2 = improve_definition(class_name, definition, prompting='', use="DeepSeek")
    print("DeepSeek Semantics: \n", semantic2)
    semantic3 = improve_definition(class_name, definition, prompting='', use="Groq")
    print("Groq Semantics: \n", semantic3)
    print("\n")
    break
```

# Classify using the LLM

```python
import os
import pathlib
from google import genai
from google.genai import types
from tqdm import tqdm
import pandas as pd

# Initialize Gemini client
client = genai.Client(api_key=GENAI_KEY)

class_labels = [
    'baby', 'bed', 'bicycle', 'chimpanzee', 'fox', 'leopard', 'man',
    'pickup_truck', 'plain', 'poppy', 'rocket', 'rose', 'snail',
    'sweet_pepper', 'table', 'telephone', 'wardrobe', 'whale', 'woman',
    'worm'
]

# Base prompt
label_string = ", ".join(class_labels)
base_prompt = f"This image belongs to one of the following categories: {label_string}. Which one is it? Just return the most appropriate category name."

root_folder = '/Users/jdieguez/Documents/master/TFM/code/cifar-fs/test/'
results = []
for subdir, _, files in os.walk(root_folder):
    for file in tqdm(files, desc="Classifying images"):
        if file.lower().endswith(('.png', '.jpg', '.jpeg')):
            image_path = os.path.join(subdir, file)
            real_class = os.path.basename(os.path.dirname(image_path))  # Folder name as ground truth

            try:
                # Prepare image
                b64_image = types.Part.from_bytes(
                    data=pathlib.Path(image_path).read_bytes(),
                    mime_type="image/jpeg"
                )

                # Query Gemini
                response = client.models.generate_content(
                    #model="gemini-2.0-flash",
                    model="gemini-2.0-flash-exp:free",
                    contents=[base_prompt, b64_image]
                )

                predicted_label = response.text.strip()

                # Save result
                results.append({
                    "file": image_path,
                    "real_class": real_class,
                    "predicted_class": predicted_label
                })

            except Exception as e:
                print(f"Error processing {image_path}: {e}")

# Save to CSV
df = pd.DataFrame(results)
df.to_csv("image_classification_results_with_real_labels.csv", index=False)
print("Done! Results saved to image_classification_results_with_real_labels.csv.")
```

```python
# trying with groq

import os
import base64
import pandas as pd
from groq import Groq
from PIL import Image
from io import BytesIO

# Classes to classify into
class_list = [
    'baby', 'bed', 'bicycle', 'chimpanzee', 'fox', 'leopard', 'man',
    'pickup_truck', 'plain', 'poppy', 'rocket', 'rose', 'snail',
    'sweet_pepper', 'table', 'telephone', 'wardrobe', 'whale', 'woman',
    'worm'
]

# Helper function to convert image to base64
def encode_image(image_path):
    with open(image_path, "rb") as img_file:
        return base64.b64encode(img_file.read()).decode("utf-8")

# Initialize Groq client
client = Groq(api_key=GROQ_KEY)

root_dir = "/Users/jdieguez/Documents/master/TFM/code/cifar-fs/test/"
results = []
i = 0
for class_name in os.listdir(root_dir):
    class_folder = os.path.join(root_dir, class_name)
    if not os.path.isdir(class_folder):
        continue

    for img_name in os.listdir(class_folder):
        img_path = os.path.join(class_folder, img_name)
        try:
            base64_image = encode_image(img_path)

            prompt = (
                f"This image belongs to one of the following classes:\n\n"
                f"{', '.join(class_list)}.\n\n"
                f"Please classify it by returning only the most appropriate class name from the list."
            )

            response = client.chat.completions.create(
                messages=[
                    {
                        "role": "user",
                        "content": [
                            {"type": "text", "text": prompt},
                            {
                                "type": "image_url",
                                "image_url": {
                                    "url": f"data:image/jpeg;base64,{base64_image}",
                                },
                            },
                        ],
                    }
                ],
                model="meta-llama/llama-4-scout-17b-16e-instruct",
            )

            pred = response.choices[0].message.content.strip()
            results.append({
                "real_class": class_name,
                "predicted_class": pred,
                "image": img_name
            })

            i += 1
            print(f"{i}. Real: {class_name}, pred: {pred}")

        except Exception as e:
            print(f"Error with {img_path}: {e}")

# Convert results to DataFrame
df = pd.DataFrame(results)
df.to_csv("groq_classification_results.csv", index=False)

print("Classification completed. Results saved to groq_classification_results.csv")
```

# Multilevel prompt

```python
# Assumed input:
# - cls: name of the class
# - wordnet_definition: text string with WordNet definition of the class
# - confused_classes: list of top-k confused class names for `cls`
# - image_inputs: list of 5 dicts, each being: {"type": "image_url" or "image", "image": <image_bytes or url>}

from groq import Groq  # or your configured client

client = Groq()

# Stage 1: Expand WordNet with scientific and contrastive detail
messages_stage1 = [
    {
        "role": "user",
        "content": [
            {
                "type": "text",
                "text": (
                    f"Definition of the class '{cls}' from WordNet:\n"
                    f"{wordnet_definition}\n\n"
                    f"Expand this with scientific and visual information about typical appearance and attributes. "
                    f"Also, contrast this class with the following visually similar/confused classes: {', '.join(confused_classes)}.\n"
                    f"Use only one concise paragraph."
                )
            }
        ]
    }
]

# Stage 2: Ground with image-based refinement
# Will be updated after stage 1 output
def build_stage2_prompt(enhanced_description):
    return [
        {
            "role": "user",
            "content": [
                {
                    "type": "text",
                    "text": (
                        f"Here is a preliminary description of the class '{cls}':\n\n"
                        f"{enhanced_description}\n\n"
                        f"Now, refine this based on the following 5 example images from the class. "
                        f"Focus only on what is actually visible and typical in these images. "
                        f"Update the description accordingly to help an AI model recognize the class visually."
                    )
                }
            ] + image_inputs
        }
    ]

# Stage 3: Purify description (remove mentions to other class names)
# Will be updated after stage 2 output
def build_stage3_prompt(refined_description):
    return [
        {
            "role": "user",
            "content": [
                {
                    "type": "text",
                    "text": (
                        f"Here is a visual description of the class '{cls}':\n\n"
                        f"{refined_description}\n\n"
                        f"Please revise this to remove any mentions to other classes like "
                        f"{', '.join(confused_classes)}. The result should only describe the target class "
                        f"'{cls}' and not rely on comparisons to others."
                    )
                }
            ]
        }
    ]


# ---- Run all 3 stages ----

try:
    # Stage 1
    response1 = client.chat.completions.create(
        model="meta-llama/llama-4-scout-17b-16e-instruct",
        messages=messages_stage1
    )
    description_stage1 = response1.choices[0].message.content.strip()

    # Stage 2
    response2 = client.chat.completions.create(
        model="meta-llama/llama-4-scout-17b-16e-instruct",
        messages=build_stage2_prompt(description_stage1)
    )
    description_stage2 = response2.choices[0].message.content.strip()

    # Stage 3
    response3 = client.chat.completions.create(
        model="meta-llama/llama-4-scout-17b-16e-instruct",
        messages=build_stage3_prompt(description_stage2)
    )
    final_description = response3.choices[0].message.content.strip()

    class_descriptions[cls] = final_description
    print(f"✅ {cls}: Final description generated.")

except Exception as e:
    print(f"❌ Error with class {cls}: {e}")
    class_descriptions[cls] = None
```

```python
import os
import json
import base64
import random
from groq import Groq
from PIL import Image
from io import BytesIO

with open("../wn_descriptions.json", 'r') as f:
    wn = json.load(f)

with open("../most_confused.json", 'r') as f:
    mc = json.load(f)

# Dictionary to hold descriptions
class_descriptions = {}

for dataset in ['train', 'test', 'val']:
    base_path = "/Users/jdieguez/Documents/master/TFM/code/cifar-fs/" + dataset
    class_names = sorted([d for d in os.listdir(base_path) if os.path.isdir(os.path.join(base_path, d))])
    
    client = Groq(api_key=GROQ_KEY)
    
    # Helper: encode image to base64
    def encode_image(image_path):
        with open(image_path, "rb") as f:
            return base64.b64encode(f.read()).decode("utf-8")
    
    
    
    for cls in class_names:
        print(cls)
        # Get extra info
        wn_definition = wn[cls]
        confused_with = mc[cls]

        # Get the images
        class_folder = os.path.join(base_path, cls)
        image_files = [f for f in os.listdir(class_folder) if f.endswith(".png")]
        selected_images = random.sample(image_files, min(5, len(image_files)))
        image_inputs = []
        for img_name in selected_images:
            image_path = os.path.join(class_folder, img_name)
            encoded = encode_image(image_path)
            image_inputs.append({
                "type": "image_url",
                "image_url": {
                    "url": f"data:image/jpeg;base64,{encoded}"
                }
            })

        # Prompt 1:
        prompt1 = [
            {
                "role": "user",
                "content": [
                    {
                        "type": "text",
                        "text": (
                            f"Definition of the class '{cls}': {wn_definition}\n\n"
                            f"Expand this with scientific and visual information about typical appearance and attributes. "
                            f"Also, contrast this class with the following visually similar/confused classes: {', '.join(confused_with)}.\n"
                            f"Use only one concise paragraph."
                        )
                    }
                ]
            }
        ]
        response1 = client.chat.completions.create(
            model="meta-llama/llama-4-scout-17b-16e-instruct",
            messages=prompt1
        )
        description1 = response1.choices[0].message.content.strip()
        print(description1)

        # Prompt 2:
        prompt2 = [
            {
                "role": "user",
                "content": [
                    {
                        "type": "text",
                        "text": (
                            f"Here is a preliminary description of the class '{cls}':\n\n"
                            f"{description1}\n\n"
                            f"Now, refine this based on the following 5 example images from the class. "
                            f"Focus only on what is actually visible and typical in these images. "
                            f"Update the description accordingly to help an AI model recognize the class visually."
                        )
                    }
                ] + image_inputs
            }
        ]
        response2 = client.chat.completions.create(
            model="meta-llama/llama-4-scout-17b-16e-instruct",
            messages=prompt2
        )
        description2 = response2.choices[0].message.content.strip()
        print(description2)

        # Prompt 3:
        prompt3 = [
            {
                "role": "user",
                "content": [
                    {
                        "type": "text",
                        "text": (
                            f"Here is a visual description of the class '{cls}':\n\n"
                            f"{description2}\n\n"
                            f"Revise this to remove any mentions to other classes like "
                            f"{', '.join(confused_with)}. The output should be a long paragraf that only describe the target class "
                            f"'{cls}' and not rely on comparisons to others."
                        )
                    }
                ]
            }
        ]
        response3 = client.chat.completions.create(
            model="meta-llama/llama-4-scout-17b-16e-instruct",
            messages=prompt3
        )
        description3 = response3.choices[0].message.content.strip()
        print(description3)
        
        class_descriptions[cls] = description3
```

```python
# Save to JSON
with open("../multi_descriptions.json", "w") as f:
    json.dump(class_descriptions, f, indent=2)
```
