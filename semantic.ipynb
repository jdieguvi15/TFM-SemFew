```python
# Use wordnet for small definition
import nltk
from nltk.corpus import wordnet as wn
from transformers import pipeline

def get_wordnet_definition(class_name):
    synsets = wn.synsets(class_name)
    if synsets:
        return synsets[0].definition()  # TODO: Take the first definition?
    return None
```

Original Prompt:

f'"{definition}" is the definition of the {class_name}. Please rewrite and expand this definition to make it more detailed and consistent with scientific fact. Briefness is required, using only one paragraph.'

Proposed Prompt:

f'"{definition}" is the definition of the {class_name}. Please rewrite and expand this definition to make it more visually detailed and consistent with scientific fact for a IA model to differentiate it from this class to other classes like {set(labels) - set([class_name])}. Briefness is required, using only one paragraph.'

```python
from google import genai
from openai import OpenAI
from groq import Groq

def improve_definition(class_name, definition, prompting='Original', use="Gemini"):
    if prompting == 'Original':
        prompt = f'"{definition}" is the definition of the {class_name}. Please rewrite and expand this definition to make it more detailed and consistent with scientific fact. Briefness is required, using only one paragraph.'
    else:
        prompt = f'"{definition}" is the definition of the {class_name}. Please rewrite and expand this definition to make it more visually detailed and consistent with scientific fact for a IA model to differentiate it from this class to other classes like {set(labels) - set([class_name])}. Briefness is required, using only one paragraph.'

    if use == "Gemini":
        client = genai.Client(api_key="AIzaSyCyu-bIWeWKIdd2IQjZVDG9T6ZBlFdazJU")
    
        response = client.models.generate_content(
            model="gemini-2.0-flash", contents=prompt
        )
        answer = response.text
        
    elif use == "DeepSeek":
        client = OpenAI(
          base_url="https://openrouter.ai/api/v1",
          api_key="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",
        )
        
        completion = client.chat.completions.create(
          extra_body={},
          model="deepseek/deepseek-r1:free",
          messages=[
            {
              "role": "user",
              "content": prompt
            }])
        answer = completion.choices[0].message.content

    elif use == "Groq":
        client = Groq(api_key="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")
        chat_completion = client.chat.completions.create(
            messages=[
                {
                    "role": "user",
                    "content": prompt,
                }
            ],
            model="llama3-8b-8192",
        )
        answer = chat_completion.choices[0].message.content
        
    return answer
```

# Classify using the LLM

```python
import os
import pathlib
from google import genai
from google.genai import types
from tqdm import tqdm
import pandas as pd

# Initialize Gemini client
client = genai.Client(api_key="AIzaSyCyu-bIWeWKIdd2IQjZVDG9T6ZBlFdazJU")

class_labels = [
    'baby', 'bed', 'bicycle', 'chimpanzee', 'fox', 'leopard', 'man',
    'pickup_truck', 'plain', 'poppy', 'rocket', 'rose', 'snail',
    'sweet_pepper', 'table', 'telephone', 'wardrobe', 'whale', 'woman',
    'worm'
]

# Base prompt
label_string = ", ".join(class_labels)
base_prompt = f"This image belongs to one of the following categories: {label_string}. Which one is it? Just return the most appropriate category name."

root_folder = '/Users/jdieguez/Documents/master/TFM/code/cifar-fs/test/'
results = []
for subdir, _, files in os.walk(root_folder):
    for file in tqdm(files, desc="Classifying images"):
        if file.lower().endswith(('.png', '.jpg', '.jpeg')):
            image_path = os.path.join(subdir, file)
            real_class = os.path.basename(os.path.dirname(image_path))  # Folder name as ground truth

            try:
                # Prepare image
                b64_image = types.Part.from_bytes(
                    data=pathlib.Path(image_path).read_bytes(),
                    mime_type="image/jpeg"
                )

                # Query Gemini
                response = client.models.generate_content(
                    #model="gemini-2.0-flash",
                    model="gemini-2.0-flash-exp:free",
                    contents=[base_prompt, b64_image]
                )

                predicted_label = response.text.strip()

                # Save result
                results.append({
                    "file": image_path,
                    "real_class": real_class,
                    "predicted_class": predicted_label
                })

            except Exception as e:
                print(f"Error processing {image_path}: {e}")

# Save to CSV
df = pd.DataFrame(results)
df.to_csv("image_classification_results_with_real_labels.csv", index=False)
print("Done! Results saved to image_classification_results_with_real_labels.csv.")
```
